# Data Analysis and Visualization

```python
cancelled_perc = Hotel["is_canceled"].value_counts(normalize = True )
cancelled_perc
```

```python
cancelled_perc = Hotel["is_canceled"].value_counts(normalize = True )
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title("Reservation Status Count")
plt.bar([" Not Canceled", "Canceled"],Hotel["is_canceled"].value_counts() , edgecolor = "k" , width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,5))
ax1= sns.countplot(x = "hotel", hue = "is_canceled" , data = Hotel , palette = "Blues")
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.xlabel("Hotel")
plt.ylabel("Number Of Reservations")
plt.legend(["Not Canceled" , "Canceled"])
plt.show()
```

```python
resort_hotel = Hotel[Hotel["hotel"] == "Resort Hotel"]
resort_hotel ["is_canceled"].value_counts(normalize = True)
```

```python
city_hotel = Hotel[Hotel["hotel"] == "City Hotel"]
city_hotel ["is_canceled"].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel = city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title("Average Daily Rate in City and Resort Hotel" , fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel["adr"], label = "Resort Hotel")
plt.plot(city_hotel.index, city_hotel["adr"], label = "City Hotel")
plt.legend(fontsize = 20)
plt.show()
```

```python
Hotel["month"] = Hotel["reservation_status_date"].dt.month
plt.figure(figsize = (18,8))
ax1 = sns.countplot(x = "month" , hue = "is_canceled" , data = Hotel, palette = "bright")
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Resveration Status Per Month", size = 23)
plt.xlabel("Month")
plt.ylabel("Number of Reservations")
plt.legend(["Not Canceled", " Canceled"])
plt.show()
```

```python
# Average daily rate (ADR)

plt.figure(figsize = (15,8))
plt.title("ADR Per Month" , fontsize = 25)
sns.barplot(x="month",y="adr", data =Hotel[Hotel["is_canceled"]==1].groupby("month")[["adr"]].sum().reset_index())
#plt.legend(fontsize = 20)
plt.show()
```

```python
cancelled_data = Hotel[Hotel["is_canceled"] == 1]
top_10_country = cancelled_data["country"].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title("Top 10 Countries With Reservation Canceled")
plt.pie(top_10_country, autopct = "%.2f", labels = top_10_country.index)
plt.show()
```

```python
cancelled_Hotel_adr = cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
cancelled_Hotel_adr.reset_index(inplace = True)
cancelled_Hotel_adr.sort_values("reservation_status_date", inplace = True)

not_cancelled_data = Hotel[Hotel["is_canceled"] == 0]
not_cancelled_Hotel_adr = not_cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
not_cancelled_Hotel_adr.reset_index(inplace = True)
not_cancelled_Hotel_adr.sort_values("reservation_status_date" , inplace = True)

plt.figure(figsize = (20,6))
plt.title("Average Daily Rate")
plt.plot(not_cancelled_Hotel_adr["reservation_status_date"],not_cancelled_Hotel_adr["adr"], label = "not cancelled")
plt.plot(cancelled_Hotel_adr["reservation_status_date"],cancelled_Hotel_adr["adr"], label = "cancelled")
plt.legend()
```

```python
cancelled_Hotel_adr = cancelled_Hotel_adr[(cancelled_Hotel_adr["reservation_status_date"]>"2016") & (cancelled_Hotel_adr["reservation_status_date"]<"2017 -09")]
not_cancelled_Hotel_adr = not_cancelled_Hotel_adr[(not_cancelled_Hotel_adr["reservation_status_date"]>"2016") & (not_cancelled_Hotel_adr["reservation_status_date"]<"2017-09")]
```

```python
plt.figure(figsize=(20, 6))
plt.title("Average Daily Rate", fontsize = 25)
plt.plot(not_cancelled_Hotel_adr["reservation_status_date"], not_cancelled_Hotel_adr["adr"], label="not cancelled")
plt.plot(cancelled_Hotel_adr["reservation_status_date"], cancelled_Hotel_adr["adr"], label="cancelled")
plt.legend(fontsize=20)
plt.show()
```
